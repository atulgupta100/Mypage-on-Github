```python

import matplotlib.pyplot as plt

# --- 1. SETTING THE SCENE ---
hours = 24             # How long are we watching the cells?
cells = 155             # Starting amount (The "Seed")
growth_rate = 0.9      # How fast do they multiply?
max_capacity = 10    # How much food/space is in the tank?

history = []           # A notebook to write down the results

# --- 2. THE GROWTH LOOP ---
for hour in range(hours):
    history.append(cells)

    # Calculate how many new cells are born this hour
    # Logic: Growth slows down as we get closer to the max_capacity
    new_cells = cells * growth_rate * (1 - cells / max_capacity)

    cells = cells + new_cells

# --- 3. SHOW THE RESULT ---
plt.plot(history)
plt.title("My First Bio-Simulator")
plt.xlabel("Time (Hours)")
plt.ylabel("Number of Cells")
plt.show()
```
